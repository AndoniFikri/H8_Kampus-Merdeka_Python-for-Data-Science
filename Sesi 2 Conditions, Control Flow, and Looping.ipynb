### Break and Continue Statements

```python
n = 5 
while n > 0 :
    n -= 1
    if n == 2: 
        break 
    print(n)
print('Loop ended.')
```

```python
n = 5 
while n > 0 :
    n -= 1
    if n == 2: 
        continue 
    print(n)
print('Loop ended.')
```

### Else Clause

```python
n = 5 
while n >0 : 
    n -= 1 
    print(n)
else: 
    print("Loop Done.")
```

```python
n = 5 
while n > 0 :
    n -= 1
    if n == 2: 
        break 
    print(n)
else:
    print('Loop ended.')
```

### Nested While Loops

```python
age = 12 
gender = 'M'
if age < 18:
    if gender == 'M':
        print('son')
    else:
        print('daughter')
elif age >= 18 and age < 65:
    if gender == 'M':
        print('father')
    else:
        print('mother')
else:
    if gender == 'M':
        print(' grandfather')
    else:
        print(' grandmother ')
```

# Application of while and for loops

```python
#For loop Example
temp = input('Masukan temp. yang ingin doikonversi, eg 45F, 120C :')
degree = int(temp[:-1])
convertion = temp[-1]

if convertion == "C":
    result = int(round((9 * degree) / 5 + 32))
    degree = 'F'
elif convertion == "F":
    result = int(round((degree - 32) * 5 / 9))
    degree = 'C'
else:
    print("Masukan input yang benar")
    
print(f"Temperaturnya adalah {result}{degree}")
```
