# Loop and Iterators

## While

```python
# Infinite Loop 
while True: 
    print('hello')
```

```python
message = 'spam'
while len(message) > 0:
    print(message)
    message = message[:(-1)]
```

```python
a = 0 
b = 13
while a<b: 
    print(a, end=' ')
    a +=1
```

## For Loops

```python
for word in ['raka', 'billy', 'kode']: 
    print(word, end='\n')
```

```python
sum = 0 
for i in [1,2,3,4]: 
    sum += i
sum
```

```python
prod = 1 
for item in [1,2,3,4] : prod *= item 
prod
```

```python
text = 'foxy fox'
words = ('is', 'great', 'animal')
for chara in text: print(chara)
```

```python
for i in words : print(i)
```

```python
d = [(1,2), (3,4), (5,6)]

for (x,y) in d : 
    print(x,y)
```

```python
for l in 'Idaz': 
    if l =='a':
        continue
    print(l)
```

# Function and Package

## Function

```python
def add(a, b): 
    return a + b
```

```python
add(5,4)
```

```python
add_things = add(10, 4)
add_things
```

```python
def intersection(first_word, second_word): 
    result = []
    for x in first_word: 
        if x in second_word: 
            result.append(x)
    return result 
```

```python
intersection('spam', 'scam')
```
